```python
import matplotlib.pyplot as plt
import json
```

```python
hours = [1,2,3,4,5]
salary = [1,4,9,16,25]
```

```python
plt.scatter(hours, salary)

plt.title('Salary-hours', fontsize=16)
plt.xlabel("Hours")
plt.ylabel("Salary")


plt.show
```

```python
yValues = []
xValues = []

with open('unemployment.json') as file:
    filename = json.load(file)
    
    for entry in filename['data']:
        xValues.append(entry['month'])
        yValues.append(entry['rate'])
        
        
plt.plot(xValues, yValues)
plt.title("Unemployment: 2017", fontsize=18)
plt.xlabel('Month')
plt.ylabel('Rate (%)')


plt.show
```
